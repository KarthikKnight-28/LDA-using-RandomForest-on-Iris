```python
df = df.drop(df.index[0]).reset_index(drop=True)
```

```python
df['sepal-length'] = pd.to_numeric(df['sepal-length'])
df['sepal-width']=pd.to_numeric(df['petal-width'])
df['petal-length']=pd.to_numeric(df['petal-length'])
df['petal-width']=pd.to_numeric(df['petal-width'])
```

```python
x=df.iloc[:,0:4].values
```

```python
x
```

```python
sc=StandardScaler()
x=sc.fit_transform(x)
```

```python
lb=LabelEncoder()
y=lb.fit_transform(y)
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
lda=LinearDiscriminantAnalysis(n_components=2)
```

```python
x_train=lda.fit_transform(x_train,y_train)
```

```python
x_test=lda.transform(x_test)
```

```python
cf=RandomForestClassifier(max_depth=2,random_state=0)
```

```python
cf.fit(x_train,y_train)
```

```python
y_pred=cf.predict(x_test)
```

```python
print('Accuracy:' +str(accuracy_score(y_test,y_pred)))
```

```python
conf_m=confusion_matrix(y_test,y_pred)
```
